```python
import urllib2 
from bs4 import BeautifulSoup
```

```python
url = 'http://python.org'

def get_site_html(url):
    source = urllib2.urlopen(url).read()
    return source
source = get_site_html(url)
'Python' in source
```

```python
source = get_site_html(url)
tree = BeautifulSoup(source)
tree
```
